```python
globleWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globleWarming_df
```

```python
# drop unnumarical columns beacuse heatmap works on numaric data
```

```python
globleWarming_df=globleWarming_df.drop(columns=['Country Code','Indicator Name','Indicator Code'], axis=1).set_index('Country Name')
globleWarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globleWarming_df)
```

```python
# set minimum values and maximum values
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globleWarming_df, vmin =0, vmax = 21 )
```

```python
# change color hof heepmap se (cmap="xyz")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globleWarming_df, cmap = "coolwarm" )
```

```python
# to showe value of data in map use (annot = True)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globleWarming_df, cmap = "coolwarm", annot =True)
```

```python
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']])
```

```python
sns.heatmap(arr_2d, annot = annot_arr, fmt='s')  # fmt --> what is type of new array (string='s')(decimal='d')
```

```python
# style in data
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.9, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
sns.heatmap(globleWarming_df, cmap = "coolwarm", annot =True, annot_kws = annot_kws)
```

```python
# saperate box for all elements
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globleWarming_df, cmap = "coolwarm", annot =True, linewidth=4)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globleWarming_df, cmap = "coolwarm", annot =True, linewidth=4, linecolor='k')
```

```python
# hide color bar
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globleWarming_df,cbar=False)
```

```python
# remove sticks
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globleWarming_df, cbar=False,xticklabels=False, yticklabels=False)
```

```python
# now work with color bar
```

```python
plt.figure(figsize=(16,9))

cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }

sns.heatmap(globleWarming_df,cbar_kws = cbar_kws)
```

```python
# change x & y axis according to our requirements
```

```python
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(globleWarming_df, xticklabels = np.arange(0,15)  , yticklabels = country_code)
```

```python
# show clear perfect
```

```python
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(globleWarming_df)
 
ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name",)
 
sns.set(font_scale=3) # set fontsize 2
```
