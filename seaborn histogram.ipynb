# Histogram / eaborn distplot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df
```

```python
sns.displot(tips_df["tip"])   # two new version
```

```python
sns.histplot(tips_df["tip"]) # two new version
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["size"], bins=100)
```

```python
sns.distplot(tips_df["total_bill"], color='r')
```

```python
sns.distplot(tips_df["total_bill"], axlabel="Total Bill")
```

```python
sns.distplot(tips_df["total_bill"], label="Total bill")
plt.legend()
plt.title="Histogram of Total bill"
plt.show()
```

```python
plt.figure(figsize=(16,9))

sns.set()
sns.distplot(tips_df["total_bill"], label="Total bill")
plt.legend()
plt.title="Histogram of Total bill"
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))

sns.set()
sns.distplot(tips_df["total_bill"], bins=bins) #label="Total bill"
plt.xticks(bins)
#plt.legend()
plt.title="Histogram of Total bill"
plt.show()
```
