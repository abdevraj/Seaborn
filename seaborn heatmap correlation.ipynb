```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
globleWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globleWarming_df
```

```python
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(globleWarming_df.corr(), annot = True, linewidth=3)
ax.tick_params(size=10, color='k',labelsize=10, labelcolor='k')

plt.title=("Who is responsible for flobelwarming")
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset =load_breast_cancer()
cancer_dataset
```

```python
#convert this data into DataFrame
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                        columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax = sns.heatmap(cancer_df.corr(), annot=True, linewidth=3)
ax.tick_params(size=10, color='k', labelsize=10, labelcolor='k')
plt.title=("Heatmap of Breast Cancer Patients")
plt.show()
```
